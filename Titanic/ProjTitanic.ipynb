```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score

tabela = pd.read_csv("titanic.csv")


#1ªclasse-> mais cara, 3ªclasse->2º mais cara,2ºclasse->mais barata
#1->sim,0->não
#male=1 e female=0

tabela.loc[tabela['Sex']=='male','numSex'] = 1
tabela.loc[tabela['Sex']=='female','numSex'] = 0

tabela = tabela.drop(['Sex'],axis=1)
#display(tabela)

#sns.heatmap(tabela[['Survived','numSex','Pclass']].corr(),annot=True, cmap = "Wistia")
#Sexo e tipo de classe não determinaram a sobrevivência(Não estão correlacionados)

#ML

y = tabela["Survived"]
x = tabela[["numSex","Pclass"]]

x_treino,x_teste,y_treino,y_teste = train_test_split(x,y,test_size=0.3)

#modelo_regressaolinear = LinearRegression()
modelo_arvoredecisao = RandomForestRegressor()

#modelo_regressaolinear.fit(x_treino,y_treino)
modelo_arvoredecisao.fit(x_treino,y_treino)

#previsao_regressaolinear = modelo_regressaolinear.predict(x_teste)
previsao_arvoredecisao = modelo_arvoredecisao.predict(x_teste)

#print(r2_score(y_teste, previsao_regressaolinear))
#print(r2_score(y_teste, previsao_arvoredecisao))

#Calcule a probabilidade condicional de que uma pessoa sobreviva, dado seu sexo e classe de passageiro:


aux = pd.DataFrame()

aux = tabela[['Survived','numSex','Pclass']].loc[tabela['Survived']==1]
aux.drop(['Survived'],axis=1,inplace=True)
#aux['numSex'] = aux['numSex'].loc[aux['numSex']==1]
#aux['Pclass'] = aux['Pclass'].loc[aux['Pclass']==3]
#aux= aux.dropna(axis=0)

#aux.shape[0]

#display(aux)

previsao = modelo_arvoredecisao.predict(aux)

aux['prev'] = previsao

#isplay(aux)

#aux.loc[0:40].plot(kind='bar', x='numSex', y='prev', figsize=(15,6), colormap='prism');

#109 homens e 233 mulheres sobreviveram ->342 no total 
#91 mulheres(96%) e 45 homens(37%) eram da 1ªclasse
#70 mulheres(48-51%) e 17 homens(15%) eram da 2ªclasse
#72 mulheres(91-93%) e  47 homens(14%) eram da 3ªclasse

#Aumentou a probabilidade de sobreviver: Ser mulher, e para homens, maior classe.

#Qual é a probabilidade de uma criança que está na terceira classe e tem 10 anos ou menos sobreviver? 


y = tabela["Survived"]
x = tabela[["Age","Pclass"]]

x_treino,x_teste,y_treino,y_teste = train_test_split(x,y,test_size=0.1)

#modelo_regressaolinear = LinearRegression()
modelo_arvoredecisao = RandomForestRegressor()

#modelo_regressaolinear.fit(x_treino,y_treino)
modelo_arvoredecisao.fit(x_treino,y_treino)

#previsao_regressaolinear = modelo_regressaolinear.predict(x_teste)
previsao_arvoredecisao = modelo_arvoredecisao.predict(x_teste)



dfB = pd.DataFrame()

dfB = tabela[['Survived','Age','Pclass']].loc[tabela['Survived']==1]
dfB['Age'] = dfB['Age'].loc[dfB['Age']<=10]
dfB['Pclass'] = dfB['Pclass'].loc[dfB['Pclass']==3]

dfB.drop(['Survived'],axis=1,inplace=True)
dfB= dfB.dropna(axis=0)

#display(dfB)

previsao = modelo_arvoredecisao.predict(dfB)
#print(previsao)

dfB['prev'] = previsao
#display(dfB)

#dfB.plot(kind='bar', x='Age', y='prev', figsize=(15,6), colormap='prism');

#Crianças recém nascidas tinham boa probabilidade de sobreviver(>=90%).
#maiores que um até três anos pouca proba(<50%).
#Mais velhas 7 até 10 anos, pouca também(<40%).
#Entre 4 à 6 anos, uma proba média(60%).


#C- Quanto as pessoas pagaram para estar no navio? Calcule a expectativa de tarifa condicionada por classe de passageiro:

#tabela['Fare'] = tabela[['Fare']].loc[tabela['Pclass']==3]
#tabela = tabela.dropna(axis=0)
#media = tabela['Fare'].mean()
#display(tabela)
#print(f"A média de tarifa paga pela terceira classe é {media:.5f}")

#A média de tarifa paga pela primeira classe foi 84.15469
#A média de tarifa paga pela segunda classe foi 20.66218
#A média de tarifa paga pela terceira classe foi 13.70771
```
